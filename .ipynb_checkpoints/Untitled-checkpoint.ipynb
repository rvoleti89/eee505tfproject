```python
import numpy as np
import soundfile as sf
import scipy.signal as signal
import matplotlib.pyplot as plt
import matplotlib.style as ms
ms.use('seaborn-muted')
%matplotlib inline

# and IPython.display for audio output
import IPython.display

# Librosa for audio
import librosa
# And the display module for visualization
import librosa.display
import seaborn as sns

# import pydrobert speech library
import pydrobert.speech
```

```python
data, samplerate = sf.read("LibriSpeech/train-clean-100/78/368/78-368-0000.flac")

data
```

```python
time = np.linspace(0, len(data)/samplerate, len(data))
sns.set_style('whitegrid')
sns.set_context('paper', font_scale=2.2)
sns.set_style({'font.family': 'serif'})
sns.despine()

plt.figure(figsize=(12, 6))

plt.plot(time, data)

plt.ylabel('Signal Amplitude')
plt.xlabel('Time [sec]')
plt.title('Raw Speech Signal from LibriSpeech Sample')

plt.savefig('raw.pdf', format='pdf', dpi=1200)
```

```python
freq, time, Sxx = signal.spectrogram(data, samplerate, scaling='spectrum')
plt.pcolormesh(time, freq, Sxx)

# add axis labels
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
```

```python
from tftb.generators import amgauss, fmlin
z = amgauss(128, 50, 40) * fmlin(128, 0.05, 0.3, 50)[0]
```

```python
from tftb.processing import WignerVilleDistribution

wvd = WignerVilleDistribution(z)
wvd.run()
wvd.plot(kind='contour')
```

```python
sns.set_context('paper', font_scale=1.5)
sns.set_style({'font.family': 'serif'})
sns.despine()

S = librosa.feature.melspectrogram(data, sr=samplerate, n_mels=128)

# Convert to log scale (dB). We'll use the peak power (max) as reference.
log_S = librosa.power_to_db(S, ref=np.max)

# Make a new figure
plt.figure(figsize=(12,4))

```

```python
mfcc        = librosa.feature.mfcc(S=log_S, n_mfcc=20, sr=samplerate)

sns.set_context('paper', font_scale=1.5)


# How do they look?  We'll show each in its own subplot
plt.figure(figsize=(12, 6))

plt.subplot(2,1,1)
# Display the spectrogram on a mel scale
# sample rate and hop length parameters are used to render the time axis
librosa.display.specshow(log_S, sr=samplerate, x_axis='time', y_axis='mel')

# Put a descriptive title on the plot
plt.title('Mel Scale Power Spectrogram')

# draw a color bar
plt.colorbar(format='%+02.0f dB')

plt.subplot(2,1,2)
librosa.display.specshow(mfcc)
plt.ylabel('MFCC')
plt.colorbar()

plt.tight_layout()

plt.savefig('mel.pdf', format='pdf', dpi=1200)
```
