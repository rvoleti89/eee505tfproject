```python
sns.set_context('paper', font_scale=1.5)
sns.set_style({'font.family': 'serif'})
sns.despine()

S = librosa.feature.melspectrogram(data, sr=samplerate, n_mels=128)

# Convert to log scale (dB). We'll use the peak power (max) as reference.
log_S = librosa.power_to_db(S, ref=np.max)

# Make a new figure
plt.figure(figsize=(12,4))

```

```python
mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=20, sr=samplerate)

sns.set_context('paper', font_scale=1.5)

plt.figure(figsize=(12, 6))

# Display the spectrogram on a mel scale
# sample rate and hop length parameters are used to render the time axis
librosa.display.specshow(log_S, sr=samplerate, x_axis='time', y_axis='linear')

# Put a descriptive title on the plot
plt.title('Linear Scale (in Frequency) Power Spectrogram')
plt.savefig('lin_spec.pdf', format='pdf', dpi=1200)
```

```python
mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=20, sr=samplerate)

sns.set_context('paper', font_scale=1.5)

plt.figure(figsize=(12, 6))

# Display the spectrogram on a mel scale
# sample rate and hop length parameters are used to render the time axis
librosa.display.specshow(log_S, sr=samplerate, x_axis='time', y_axis='mel')

# Put a descriptive title on the plot
plt.title('Mel Scale (in Frequency) Power Spectrogram')
plt.savefig('mel_spec.pdf', format='pdf', dpi=1200)# draw a color bar
#plt.colorbar(format='%+02.0f dB')
```

```python
sns.set_context('paper', font_scale=1.5)
plt.figure(figsize=(12, 6))
librosa.display.specshow(mfcc, x_axis='time')

plt.ylabel('MFCC')
plt.colorbar()

plt.tight_layout()

plt.savefig('mfcc.pdf', format='pdf', dpi=1200)
```

```python
sns.set_context('paper', font_scale=1.5)
plt.figure(figsize=(12, 6))
cqt = np.abs(librosa.core.cqt(data, sr=samplerate))

librosa.display.specshow(librosa.amplitude_to_db(cqt, ref=np.max), sr=samplerate, x_axis='time', y_axis='cqt_note')
plt.colorbar(format='%+2.0f dB')
plt.title('Constant-Q power spectrum')
plt.tight_layout()
plt.savefig('cqt.pdf', format='pdf', dpi=1200)
```
